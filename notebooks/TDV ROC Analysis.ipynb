```python
import os
import shutil

def rename_snr_folders_to_snr0(directory_path):
    """
    Rename all SNR folders to SNR0 with sequential row numbers to avoid duplication
    """
    # Get all folder names in the directory
    folder_names = [f for f in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, f)) and f != '.DS_Store']
    
    # Filter folders that contain SNR
    snr_folders = [f for f in folder_names if 'SNR' in f]
    
    print(f"Found {len(snr_folders)} SNR folders to rename:")
    for folder in snr_folders:
        print(f"  - {folder}")
    
    # Sort folders for consistent ordering
    snr_folders.sort()
    
    # Rename each folder to SNR0 with sequential row numbers
    for i, old_folder_name in enumerate(snr_folders, 1):
        # Create new folder name with SNR0 and sequential row number
        new_folder_name = f"koi-KOI_SNR0_Row{i}"
        
        old_path = os.path.join(directory_path, old_folder_name)
        new_path = os.path.join(directory_path, new_folder_name)
        
        try:
            # Rename the folder
            os.rename(old_path, new_path)
            print(f"Renamed: {old_folder_name} → {new_folder_name}")
        except Exception as e:
            print(f"Error renaming {old_folder_name}: {e}")
    
    print(f"\nCompleted renaming {len(snr_folders)} folders to SNR0 with sequential row numbers.")

# Execute the renaming
directory_path = '../data/Output_data'
rename_snr_folders_to_snr0(directory_path)
```

```python
import os
import json
import pandas as pd

def _read_roc_metrics(folder_path):
    """Return {'db_over_dt_z_score','t_score','t_score_t14'} for a KOI output folder.
    Prefers the new tdv_metrics_*.json; falls back to the legacy parameters_final*.csv."""
    for file in os.listdir(folder_path):
        if file.startswith('tdv_metrics') and file.endswith('.json'):
            with open(os.path.join(folder_path, file)) as f:
                j = json.load(f)
            return {'db_over_dt_z_score': j.get('db_dt_global_zscore'),
                    't_score': j.get('t_score_b'),
                    't_score_t14': j.get('t_score_t14')}
    for file in os.listdir(folder_path):
        if file.startswith('parameters_final') and file.endswith('.csv'):
            df = pd.read_csv(os.path.join(folder_path, file))
            def g(name):
                return df[df['Parameter'] == name]['Best Params'].values[0] if name in df['Parameter'].values else None
            return {'db_over_dt_z_score': g('db_over_dt_z_score'),
                    't_score': g('t_score'),
                    't_score_t14': g('t_score_t14')}
    return None

def extract_values_from_csv_for_snr(root_dir, snr_values):
    results = {}
    # Loop through each directory and subdirectory in the root directory
    for root, dirs, files in os.walk(root_dir):
        for folder_name in dirs:
            # Determine the specific SNR value for the current folder
            current_snr = next((snr for snr in snr_values if snr in folder_name), None)
            if current_snr:
                metrics = _read_roc_metrics(os.path.join(root, folder_name))
                if metrics is not None:
                    results[f"{folder_name}"] = metrics
    return results

# Replace '../data/Output_data' with the actual path to your directory
root_directory = '../data/Output_data_signal'
# Define the SNR values you want to include
snr_values = ['SNR0', 'SNR10', 'SNR20', 'SNR30', 'SNR50', 'SNR100']
values_dict = extract_values_from_csv_for_snr(root_directory, snr_values)
filtered_dict = {k: v for k, v in values_dict.items() if None not in v.values()}

print(filtered_dict)
```

```python
# Initialize dictionaries to keep track of counts and total occurrences with all SNR values included
db_over_dt_z_score_greater_than_3 = {'SNR0': 0, 'SNR10': 0, 'SNR20': 0, 'SNR30': 0, 'SNR50': 0, 'SNR100': 0}
t_score_greater_than_3 = {'SNR0': 0, 'SNR10': 0, 'SNR20': 0, 'SNR30': 0, 'SNR50': 0, 'SNR100': 0}
t_score_t14_greater_than_3 = {'SNR0': 0, 'SNR10': 0, 'SNR20': 0, 'SNR30': 0, 'SNR50': 0, 'SNR100': 0}

# Total counts for each SNR category
snr_total = {'SNR0': 0, 'SNR10': 0, 'SNR20': 0, 'SNR30': 0, 'SNR50': 0, 'SNR100': 0}

for key, value in filtered_dict.items():
    # Skip entries with None values
    if None in value.values():
        continue
    
    # Determine the SNR category from the key (check longer strings first to avoid substring matches)
    snr_category = None
    if 'SNR100' in key:
        snr_category = 'SNR100'
    elif 'SNR50' in key:
        snr_category = 'SNR50'
    elif 'SNR30' in key:
        snr_category = 'SNR30'
    elif 'SNR20' in key:
        snr_category = 'SNR20'
    elif 'SNR10' in key:
        snr_category = 'SNR10'
    elif 'SNR0' in key:
        snr_category = 'SNR0'
    if snr_category:
        # Update total count for this SNR category
        snr_total[snr_category] += 1
        
        # Update counts for absolute values greater than 3
        if abs(value['db_over_dt_z_score']) > 3:
            db_over_dt_z_score_greater_than_3[snr_category] += 1
        if abs(value['t_score']) > 3:
            t_score_greater_than_3[snr_category] += 1
        if abs(value['t_score_t14']) > 3:
            t_score_t14_greater_than_3[snr_category] += 1

# Calculate the relative proportions (avoid division by zero)
db_over_dt_z_score_proportions = {k: v / snr_total[k] if snr_total[k] > 0 else 0 for k, v in db_over_dt_z_score_greater_than_3.items()}
t_score_proportions = {k: v / snr_total[k] if snr_total[k] > 0 else 0 for k, v in t_score_greater_than_3.items()}
t_score_t14_proportions = {k: v / snr_total[k] if snr_total[k] > 0 else 0 for k, v in t_score_t14_greater_than_3.items()}

# Print the results
print("Proportions where |db_over_dt_z_score| > 3:")
print(db_over_dt_z_score_proportions)
print("\nProportions where |t_score| > 3:")
print(t_score_proportions)
print("\nProportions where |t_score_t14| > 3:")
print(t_score_t14_proportions)
print("\nTotal counts per SNR category:")
print(snr_total)

```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import AutoMinorLocator

# Set font and text parameters
plt.rcParams["font.family"] = "Times New Roman"
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Times New Roman'
matplotlib.rcParams['mathtext.it'] = 'Times New Roman:italic'

# Function to set ticks and axis styling, including boundary color adjustments
def ticksetax(ax, labelsize=25, ticksize=12, tickwidth=1.5, ticklength=5):
    ax.tick_params(direction='in', which='both', width=tickwidth, colors='k', 
                   bottom=True, top=True, left=True, right=True, labelsize=labelsize)
    ax.spines['bottom'].set_linewidth(tickwidth)
    ax.spines['left'].set_linewidth(tickwidth)
    ax.spines['right'].set_linewidth(tickwidth)
    ax.spines['top'].set_linewidth(tickwidth)
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['right'].set_color('black')
    ax.spines['top'].set_color('black')
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())

# Prepare data for ROC curves using all LLR categories
y_true = []
y_scores_db = []
y_scores_t_score = []
y_scores_t_score_t14 = []

for key, value in filtered_dict.items():
    # SNR6 and SNR10 are considered no signal (0), others are signal (1)
    if 'SNR0' in key:
        y_true.append(0)  # No signal
    else:  # SNR20, SNR30, SNR50, SNR100
        y_true.append(1)  # Signal
    
    # Extract the scores for ROC analysis
    y_scores_db.append(value['db_over_dt_z_score'])
    y_scores_t_score.append(value['t_score'])
    y_scores_t_score_t14.append(value['t_score_t14'])

# Calculate ROC curve and AUC for each parameter
fpr_db, tpr_db, _ = roc_curve(y_true, y_scores_db)
roc_auc_db = auc(fpr_db, tpr_db)

fpr_t_score, tpr_t_score, _ = roc_curve(y_true, y_scores_t_score)
roc_auc_t_score = auc(fpr_t_score, tpr_t_score)

# For t_score_t14, we need to flip the sign since more negative values indicate stronger signal
y_scores_t_score_t14_flipped = [-x for x in y_scores_t_score_t14]
fpr_t_score_t14, tpr_t_score_t14, _ = roc_curve(y_true, y_scores_t_score_t14_flipped)
roc_auc_t_score_t14 = auc(fpr_t_score_t14, tpr_t_score_t14)

# Plot all ROC curves on the same graph with step-wise style
fig, ax = plt.subplots(figsize=(8, 6))
ticksetax(ax, labelsize=25)  # Set label size for all tick labels to 25

plt.plot(fpr_db, tpr_db, label=r'SIPVA Fit Impact Parameter (Area = %.2f)' % roc_auc_db, color='blue', lw=2, drawstyle='steps-post')
plt.plot(fpr_t_score, tpr_t_score, label=r'$\it{Individual\ Fit}$ Impact Parameter (Area = %.2f)' % roc_auc_t_score, color='green', lw=2, drawstyle='steps-post')
plt.plot(fpr_t_score_t14, tpr_t_score_t14, label=r'$\it{Individual\ Fit}$ TDV (Area = %.2f)' % roc_auc_t_score_t14, color='red', lw=2, drawstyle='steps-post')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])  # Set the y-axis limit to (0, 1)
plt.xlabel('False Positive Rate', fontsize=25)
plt.ylabel('True Positive Rate', fontsize=25)
plt.legend(loc="lower right", fontsize=18)

# Apply custom axis ticks and styling
ticksetax(ax, labelsize=25)  # Ensure all ticks are set to font size 25
plt.tight_layout()

plt.savefig('roc_curve_updated.png', dpi=300, bbox_inches='tight')

# Remove inner grid lines
ax.grid(False)

plt.show()
```

```python
from data import get_koi

# Function to get the number of planets for each KOI
def get_koi_planet_count(koi_list):
    koi_planet_counts = {}
    for koi in koi_list:
        try:
            koi_data = get_koi(koi)
            # Access the koi_count attribute for the number of planets
            planet_count = koi_data.koi_count
            koi_planet_counts[koi] = planet_count
        except Exception as e:
            # If there's an error, print it and continue
            print(f"Error fetching KOI-{koi}: {e}")
            koi_planet_counts[koi] = None  # Set None if there's an issue fetching
    return koi_planet_counts

# Function to group KOIs by the number of planets in the system
def group_koi_by_planet_count(koi_planet_counts):
    grouped_counts = {}
    for koi, planet_count in koi_planet_counts.items():
        if planet_count is not None:
            if planet_count in grouped_counts:
                grouped_counts[planet_count].append(koi)
            else:
                grouped_counts[planet_count] = [koi]
    return grouped_counts

# List of KOIs to fetch planet counts for
merged_koi_floats = koi_list
# Get planet counts for each KOI in the list
koi_planet_counts = get_koi_planet_count(merged_koi_floats)

# Group KOIs by number of planets
grouped_koi_by_planet = group_koi_by_planet_count(koi_planet_counts)

# Print the results
for planet_count, koi_list in grouped_koi_by_planet.items():
    print(f"{len(koi_list)} KOIs belong to systems with {planet_count} planets.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from detectability import snr_square  # Ensure this is your actual SNR function

# Set font and text parameters as provided by your mentor
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'

# Function to set ticks and axis styling
def ticksetax(ax, labelsize=15, ticksize=12, tickwidth=1.5, ticklength=5):
    ax.tick_params(direction='in', which='both', width=tickwidth, colors='k', 
                   bottom=True, top=True, left=True, right=True, labelsize=labelsize)
    ax.spines['bottom'].set_linewidth(tickwidth)
    ax.spines['left'].set_linewidth(tickwidth)
    ax.spines['right'].set_linewidth(tickwidth)
    ax.spines['top'].set_linewidth(tickwidth)
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())

# Noise levels in ppm, with transformation to account for short cadence
noise_levels_ppm = [40, 200, 600]
noise_levels_flux = [level * 1e-6 for level in noise_levels_ppm]  # Convert ppm to relative flux and apply np.sqrt(15)

# Range for DB_OVER_DT (from 10^-4 to 10^-1 on a log scale)
DB_OVER_DT_values = np.logspace(-4, -1, num=100)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

for sigma_sc, noise_ppm in zip(noise_levels_flux, noise_levels_ppm):
    # Initialize list to hold SNR values for current noise level
    SNR_list = []
    
    # Compute SNR for each DB_OVER_DT value
    for dt in DB_OVER_DT_values:
        try:
            # Calculate SNR^2 (SNR squared) using snr_square function
            SNR_sq = SNR_square_LD(
                b=b,
                p=p,
                PERIOD=PERIOD,
                DB_OVER_DT=dt,
                RS_OVER_A=RS_OVER_A,
                NUM_TRANSITS=NUM_TRANSITS,
                SIGMA_SC=sigma_sc
            )
            # Ensure SNR_sq is non-negative before taking sqrt
            if SNR_sq < 0:
                SNR_sq = 0
           
            SNR_list.append(SNR_sq)
        except Exception as e:
            # Handle potential errors in snr_square
            print(f"Error computing SNR for dt={dt}: {e}")
            SNR_list.append(np.nan)  # Use NaN for undefined values
    
    # Convert SNR_list to numpy array
    SNR_array = np.array(SNR_list)
    
    # Plot only valid (non-NaN) values
    valid_indices = ~np.isnan(SNR_array)
    ax.plot(DB_OVER_DT_values[valid_indices], SNR_array[valid_indices], label=f'{noise_ppm} ppm')

# Define x range for the dashed lines
x_range = np.logspace(-4, -1, num=100)  # From 10^-4 to 10^-1

# Add horizontal dashed lines for SNR = 20 and SNR = 50 up to x = 0.1
ax.plot(x_range, [20] * len(x_range), color='black', linestyle='--', linewidth=1)
ax.plot(x_range, [50] * len(x_range), color='black', linestyle='-.', linewidth=1)

# Add text labels for SNR = 20 and SNR = 50 lines (on the left, above the line)
ax.text(1.1e-4, 25, 'LLR = 20', verticalalignment='center', color='black', fontsize=18)
ax.text(1.1e-4, 65, 'LLR = 50', verticalalignment='center', color='black', fontsize=18)

# Set axis scales and labels
ax.set_xscale('log')  # Log scale for x-axis
ax.set_yscale('log')  # Log scale for y-axis
ax.set_xlabel(r'$\dot{b}$ (Impact Parameter Rate of Change)', fontsize=25)
ax.set_ylabel('Log Likelihood Ratio (LLR)', fontsize=25)

# Set ticks and axis styles using the ticksetax function
ticksetax(ax, labelsize=20)

# Customize the legend with larger font size
ax.legend(title='Noise Level', fontsize=18, title_fontsize=20)

# **Remove Inner Grid Lines and Keep Boundary Spines**
# Disable grid lines inside the plot
ax.grid(False)  # Disables all grid lines

# Set the color of all spines to black to emphasize the boundary
for spine in ax.spines.values():
    spine.set_color('black')

# Optionally, adjust spine linewidth for better visibility
for spine in ax.spines.values():
    spine.set_linewidth(1.5)

# Final layout adjustments
plt.tight_layout()
plt.savefig('SNR_vs_DB_OVER_DT.png', dpi=300)
plt.show()
```

```python
import os
import json
import pandas as pd

def _read_dbdt(folder_path):
    """Return (db_over_dt_estim, db_over_dt_linear_reg), both in 1/yr, for a KOI output folder.
    Prefers tdv_metrics_*.json (db_dt_linreg already per-year); falls back to the legacy
    parameters_final*.csv (linear-reg slope stored per-day -> multiply by 365)."""
    for file in os.listdir(folder_path):
        if file.startswith('tdv_metrics') and file.endswith('.json'):
            with open(os.path.join(folder_path, file)) as f:
                j = json.load(f)
            return j.get('db_dt_global'), j.get('db_dt_linreg')  # both already 1/yr
    for file in os.listdir(folder_path):
        if file.startswith('parameters_final') and file.endswith('.csv'):
            df = pd.read_csv(os.path.join(folder_path, file))
            def g(name):
                return df[df['Parameter'] == name]['Best Params'].values[0] if name in df['Parameter'].values else None
            estim = g('db_over_dt_estim')
            linreg = g('db_over_dt_linear_reg')
            if linreg is not None:
                linreg *= 365  # legacy CSV stores the linear-reg slope per-day
            return estim, linreg
    return None, None

def extract_values_from_csv_for_snr(root_dir, snr_values):
    sipva_results = {}
    individual_fit_results = {}
    
    # Loop through each directory and subdirectory in the root directory
    for root, dirs, files in os.walk(root_dir):
        for folder_name in dirs:
            # Determine the specific SNR value for the current folder (check longer strings first to avoid substring matches)
            current_snr = None
            if 'SNR100' in folder_name:
                current_snr = 'SNR100'
            elif 'SNR50' in folder_name:
                current_snr = 'SNR50'
            elif 'SNR30' in folder_name:
                current_snr = 'SNR30'
            elif 'SNR20' in folder_name:
                current_snr = 'SNR20'
            elif 'SNR10' in folder_name:
                current_snr = 'SNR10'
            elif 'SNR6' in folder_name:
                current_snr = 'SNR6'
            if current_snr:
                db_over_dt_estim, db_over_dt_linear_reg = _read_dbdt(os.path.join(root, folder_name))
                # Save the extracted values into separate dictionaries
                if db_over_dt_estim is not None:
                    sipva_results[f"{folder_name}"] = {'db_over_dt_estim': db_over_dt_estim}
                if db_over_dt_linear_reg is not None:
                    individual_fit_results[f"{folder_name}"] = {'db_over_dt_linear_reg': db_over_dt_linear_reg}
    
    return sipva_results, individual_fit_results

# Replace '../data/Output_data' with the actual path to your directory
root_directory = '../data/Output_data_signal'
# Define the SNR values you want to include (all LLR categories)
snr_values = ['SNR6', 'SNR10', 'SNR20', 'SNR30', 'SNR50', 'SNR100']

# Extract the results into two separate dictionaries
sipva_dict, individual_fit_dict = extract_values_from_csv_for_snr(root_directory, snr_values)

# Print or use the extracted dictionaries
print("SIPVA Results:")
print(sipva_dict)

print("\nIndividual Fit Results:")
print(individual_fit_dict)
```

```python
def compute_absolute_differences(csv_dir, sipva_dict, individual_fit_dict, snr_values):
    """
    Computes the absolute differences between true and estimated \dot{b} values for SIPVA and Individual Fit methods.
    
    Parameters:
    - csv_dir (str): Path to the directory containing SNR CSV files.
    - sipva_dict (dict): Dictionary containing SIPVA estimates.
    - individual_fit_dict (dict): Dictionary containing Individual Fit estimates.
    - snr_values (list): List of SNR values corresponding to LLR categories.
    
    Returns:
    - pd.DataFrame: DataFrame containing LLR, Method, and Absolute Difference.
    """
    data = {
        'LLR': [],
        'Method': [],
        'Absolute_Difference': []
    }
    
    for snr in snr_values:
        csv_filename = f"SNR_{snr}.csv"
        csv_path = os.path.join(csv_dir, csv_filename)
        
        if not os.path.isfile(csv_path):
            print(f"Warning: {csv_path} does not exist. Skipping.")
            continue
        
        # Read the CSV file
        df_true = pd.read_csv(csv_path)
        
        # Ensure 'DB_OVER_DT' column exists
        if 'DB_OVER_DT' not in df_true.columns:
            print(f"Warning: 'DB_OVER_DT' column not found in {csv_path}. Skipping.")
            continue
        
        # Iterate through each row in the CSV
        for idx, row in df_true.iterrows():
            row_number = idx + 1  # Assuming row numbering starts at 1
            true_db_over_dt = row['DB_OVER_DT']
            
            # Construct the key to access the estimates
            key = f"koi-KOI_SNR{snr}_Row{row_number}"
            
            # Extract SIPVA estimate
            sipva_estimate = sipva_dict.get(key, {}).get('db_over_dt_estim', None)
            
            # Extract Individual Fit estimate (already multiplied by 365)
            individual_fit_estimate = individual_fit_dict.get(key, {}).get('db_over_dt_linear_reg', None)
            
            # Compute absolute differences if estimates are available
            if sipva_estimate is not None:
                abs_diff_sipva = abs(true_db_over_dt - sipva_estimate)/abs(true_db_over_dt)
                data['LLR'].append(snr)
                data['Method'].append('SIPVA')
                data['Absolute_Difference'].append(abs_diff_sipva)
            
            if individual_fit_estimate is not None:
                abs_diff_individual = abs(true_db_over_dt - individual_fit_estimate)/abs(true_db_over_dt)
                data['LLR'].append(snr)
                data['Method'].append('Individual Fit')
                data['Absolute_Difference'].append(abs_diff_individual)
    
    # Convert the dictionary to a DataFrame
    df_differences = pd.DataFrame(data)
    return df_differences
```

```python
import os
import pandas as pd

def compute_statistics_table(csv_dir, sipva_dict, individual_fit_dict, snr_values):
    """
    Computes Bias, Variance, and MSE between true and estimated \dot{b} values for SIPVA and Individual Fit methods.
    
    Parameters:
    - csv_dir (str): Path to the directory containing SNR CSV files.
    - sipva_dict (dict): Dictionary containing SIPVA estimates.
    - individual_fit_dict (dict): Dictionary containing Individual Fit estimates.
    - snr_values (list): List of SNR values corresponding to LLR categories.
    
    Returns:
    - pd.DataFrame: DataFrame containing LLR, Method, Bias, Variance, and MSE.
    """
    # Initialize a list to store the results
    results = []

    for snr in snr_values:
        csv_filename = f"SNR_{snr}.csv"
        csv_path = os.path.join(csv_dir, csv_filename)
        
        if not os.path.isfile(csv_path):
            print(f"Warning: {csv_path} does not exist. Skipping.")
            continue
        
        # Read the CSV file
        df_true = pd.read_csv(csv_path)
        
        # Ensure 'DB_OVER_DT' column exists
        if 'DB_OVER_DT' not in df_true.columns:
            print(f"Warning: 'DB_OVER_DT' column not found in {csv_path}. Skipping.")
            continue
        
        # Iterate through each row in the CSV
        for idx, row in df_true.iterrows():
            row_number = idx + 1  # Assuming row numbering starts at 1
            true_db_over_dt = row['DB_OVER_DT']
            
            # Construct the key to access the estimates
            key = f"koi-KOI_SNR{snr}_Row{row_number}"
            
            # Extract SIPVA estimate
            sipva_estimate = sipva_dict.get(key, {}).get('db_over_dt_estim', None)
            
            # Extract Individual Fit estimate (already multiplied by 365)
            individual_fit_estimate = individual_fit_dict.get(key, {}).get('db_over_dt_linear_reg', None)
            
            # Compute differences if estimates are available
            if sipva_estimate is not None:
                diff_sipva = sipva_estimate - true_db_over_dt
                results.append({
                    'LLR': snr,
                    'Method': 'SIPVA',
                    'Difference': diff_sipva
                })
            
            if individual_fit_estimate is not None:
                diff_individual = individual_fit_estimate - true_db_over_dt
                results.append({
                    'LLR': snr,
                    'Method': 'Individual Fit',
                    'Difference': diff_individual
                })
    
    # Convert the list of results to a DataFrame
    df_results = pd.DataFrame(results)
    
    # Group by LLR and Method to compute statistics
    grouped = df_results.groupby(['LLR', 'Method'])
    
    # Calculate Bias, Variance, and MSE
    stats = grouped['Difference'].agg(
        Bias=lambda x: x.mean(),
        Variance=lambda x: x.var(ddof=1),  # Sample variance
        MSE=lambda x: (x ** 2).mean()
    ).reset_index()
    
    # Optionally, format the statistics for better readability
    stats['Bias (%)'] = stats['Bias'] 
    stats['Variance'] = stats['Variance']
    stats['MSE'] = stats['MSE']
    
    # Select and reorder columns for the final table
    final_table = stats[['LLR', 'Method', 'Bias (%)', 'Variance', 'MSE']]
    
    # Print the table
    print(final_table.to_string(index=False))
    
    return final_table

compute_statistics_table(csv_directory, sipva_dict, individual_fit_dict, snr_values)
```
